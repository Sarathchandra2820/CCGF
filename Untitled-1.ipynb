```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.metrics.pairwise import euclidean_distances
from scipy.sparse import csgraph

def create_similarity_matrix(data):
    """Create a similarity matrix from the dataset."""
    # Use Gaussian (RBF) kernel for similarity
    distances = euclidean_distances(data)
    sigma = np.mean(distances)  # or choose a suitable value for sigma
    similarity_matrix = np.exp(-distances**2 / (2 * sigma**2))
    return similarity_matrix

def compute_laplacian(similarity_matrix):
    """Compute the Laplacian matrix from the similarity matrix."""
    # Degree matrix
    D = np.diag(np.sum(similarity_matrix, axis=1))
    # Laplacian matrix
    L = D - similarity_matrix
    return L

def spectral_clustering(data, n_clusters):
    """Perform spectral clustering on the data."""
    # Step 1: Create similarity matrix
    similarity_matrix = create_similarity_matrix(data)

    # Step 2: Compute Laplacian matrix
    L = compute_laplacian(similarity_matrix)

    # Step 3: Compute eigenvalues and eigenvectors
    eigenvalues, eigenvectors = np.linalg.eigh(L)

    # Step 4: Select the first `n_clusters` eigenvectors
    X = eigenvectors[:, :n_clusters]

    # Step 5: Normalize the eigenvector matrix
    X_normalized = X / np.linalg.norm(X, axis=1, keepdims=True)

    # Step 6: Apply K-Means clustering
    kmeans = KMeans(n_clusters=n_clusters)
    labels = kmeans.fit_predict(X_normalized)

    return labels

# Example data (using random points for demonstration)
np.random.seed(0)
n_points = 100
data = np.random.rand(n_points, 2)

# Number of clusters
n_clusters = 3

# Perform spectral clustering
labels = spectral_clustering(data, n_clusters)

# Visualize the results
plt.scatter(data[:, 0], data[:, 1], c=labels, cmap='viridis')
plt.title('Spectral Clustering Result')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.colorbar(label='Cluster Label')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans

def compute_laplacian(adjacency_matrix):
    """Compute the Laplacian matrix from the weighted adjacency matrix."""
    # Degree matrix: sum of weights for each node
    D = np.diag(np.sum(adjacency_matrix, axis=1))
    # Laplacian matrix
    L = D - adjacency_matrix
    return L

def spectral_clustering(adjacency_matrix, n_clusters):
    """Perform spectral clustering on the given weighted adjacency matrix."""
    # Step 1: Compute Laplacian matrix
    L = compute_laplacian(adjacency_matrix)

    # Step 2: Compute eigenvalues and eigenvectors
    eigenvalues, eigenvectors = np.linalg.eigh(L)

    # Step 3: Select the first `n_clusters` eigenvectors
    X = eigenvectors[:, :n_clusters]

    # Step 4: Normalize the eigenvector matrix
    X_normalized = X / np.linalg.norm(X, axis=1, keepdims=True)

    # Step 5: Apply K-Means clustering
    kmeans = KMeans(n_clusters=n_clusters)
    labels = kmeans.fit_predict(X_normalized)

    return labels

# Example: Create a weighted adjacency matrix for testing
# Weighted adjacency matrix for a simple graph
weighted_adjacency_matrix = eri_new

# Number of clusters
n_clusters = 3

# Perform spectral clustering
labels = spectral_clustering(weighted_adjacency_matrix, n_clusters)

# Visualize the results
# Create a plot where nodes are arranged in a circle
n_nodes = weighted_adjacency_matrix.shape[0]
angles = np.linspace(0, 2 * np.pi, n_nodes, endpoint=False).tolist()

# Add the first angle to the end to close the circular layout
angles += angles[:1]

# Get positions of nodes
x = np.cos(angles)
y = np.sin(angles)

# Plot each node
plt.figure(figsize=(8, 8))
for i in range(n_nodes):
    plt.scatter(x[i], y[i], c='C' + str(labels[i]), label=f'Node {i+1}')
    plt.text(x[i] * 1.1, y[i] * 1.1, str(i+1), ha='center', va='center')

plt.title('Spectral Clustering Result with Weighted Adjacency Matrix')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.xlim(-1.5, 1.5)
plt.ylim(-1.5, 1.5)
plt.gca().set_aspect('equal', adjustable='box')
plt.grid()
plt.show()
```
